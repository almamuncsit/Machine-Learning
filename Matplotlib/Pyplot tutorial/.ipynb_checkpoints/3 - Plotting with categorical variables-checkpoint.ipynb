```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
names = [4, 5, 9]
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
import matplotlib.pyplot as plt
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(15, 5))

plt.subplot(131)
plt.bar(names, values,  width=0.3, color='g')
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values, 'm--', linewidth=4.0)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
x1 = [1, 3, 5]
y1 = [7, 8, 9]

x2 = [3, 5, 8]
y2 = [4, 8, 9]

lines = plt.plot(x1, y1, x2, y2)
# use keyword args
plt.setp(lines, color='k', linewidth=2.0)
plt.show()
```
