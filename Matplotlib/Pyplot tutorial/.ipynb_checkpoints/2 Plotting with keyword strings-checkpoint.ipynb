```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
```

```python
print(data['a'])
```

```python
print(data['b'])
```

```python
print(data['c'])
```

```python
print(data['d'])
```

```python
plt.figure(1, figsize=(15, 8))
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
