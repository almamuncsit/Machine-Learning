```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]
```

```python
plt.scatter(x, y, label="Scatter", color='k', marker='o')
plt.title("Welcome to Ploting \n Check it out")
plt.xlabel('Plot Number')
plt.ylabel('Important Var')
plt.legend()
plt.show()
```

## Marker size

```python
plt.scatter(x, y, label="Scatter", color='m', s=200, marker='o')
plt.title("Welcome to Ploting \n Check it out")
plt.xlabel('Plot Number')
plt.ylabel('Important Var')
plt.legend()
plt.show()
```

## Changing Marker style

```python
plt.scatter(x, y, label="Scatter", color='m', s=200, marker='*')
plt.title("Welcome to Ploting \n Check it out")
plt.xlabel('Plot Number')
plt.ylabel('Important Var')
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, label="Scatter", color='m', s=200, marker='2')
plt.title("Welcome to Ploting \n Check it out")
plt.xlabel('Plot Number')
plt.ylabel('Important Var')
plt.legend()
plt.show()
```
