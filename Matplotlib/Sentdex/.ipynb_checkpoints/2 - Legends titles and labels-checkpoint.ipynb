```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [5, 7, 4]
```

```python
plt.plot(x, y)
plt.show()
```

## Setting Labels

```python
plt.plot(x, y)
plt.title("Welcome to Ploting")
plt.xlabel('Plot Number')
plt.ylabel('Important Var')
plt.show()
```

## Adding line break on title

```python
plt.plot(x, y)
plt.title("Welcome to Ploting \n Check it out")
plt.xlabel('Plot Number')
plt.ylabel('Important Var')
plt.show()
```

# Adding Multiple Graph together

```python
x2 = [1, 3, 4]
y2 = [10, 14, 13]
plt.plot(x, y, label='First line')
plt.plot(x2, y2, label='Second Line')
plt.title("Welcome to Ploting \n Check it out")
plt.xlabel('Plot Number')
plt.ylabel('Important Var')
plt.legend()
plt.show()
```
