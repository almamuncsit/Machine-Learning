```python
import matplotlib.pyplot as plt
```

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']
```

```python
plt.pie(slices, labels=activities)
plt.title("Welcome to Ploting \n Check it out")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
## Setting color
```

```python
plt.pie(slices, labels=activities, colors=cols)
plt.title("Welcome to Ploting \n Check it out")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## Setting Starting Angle

```python
plt.pie(slices, 
        labels=activities, 
        colors=cols,
       startangle=90)
plt.title("Welcome to Ploting \n Check it out")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## Adding Shadow

```python
plt.pie(slices, 
        labels=activities, 
        colors=cols,
       startangle=90,
       shadow = True)
plt.title("Welcome to Ploting \n Check it out")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## exploding chart

```python
plt.pie(slices, 
        labels=activities, 
        colors=cols,
       startangle=90,
       shadow = True,
       explode=(0, 0.2, 0, 0))
plt.title("Welcome to Ploting \n Check it out")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## Adding Percentage

```python
plt.pie(slices, 
        labels=activities, 
        colors=cols,
       startangle=90,
       shadow = True,
       explode=(0, 0.2, 0, 0),
       autopct='%1.1f%%')
plt.title("Welcome to Ploting \n Check it out")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
